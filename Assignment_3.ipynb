Import Libraries & Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Housing.xlsx")

df = df.head()
print(df)
print(df.info())
```

1. Scatter Plot

    Area vs Price

```python
plt.scatter(df['area'], df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Scatter Plot: Area vs Price")
plt.show()
```

2. Bar Plot

    Prices of First 10 Houses

```python
plt.bar(df.index[:10], df['price'][:10])
plt.title("Prices of First 10 Houses")
plt.xlabel("House Index")
plt.ylabel("Price")
plt.show()
```

```python
colors = ['red','blue','green','orange','purple']
plt.figure(figsize=(6,4))
plt.bar(df.index[:10], df['price'][:10], color=colors)
plt.title("Bar Plot of First 10 Houses")
plt.xlabel("House Index")
plt.ylabel("Price")
plt.show()
```

3. Box Plot

    Price Distribution

```python
plt.figure(figsize=(6,4))

plt.boxplot(df['price'],
            patch_artist=True,
            boxprops=dict(facecolor='lightblue'),
            medianprops=dict(color='red'))

plt.yscale('log')

plt.title("Box Plot of House Prices")
plt.ylabel("Price (log scale)")
plt.show()
```

4. Pie Chart

    Furnishing Status Distribution

```python
labels = df['furnishingstatus'].value_counts().index
sizes = df['furnishingstatus'].value_counts().values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Furnishing Status Distribution")
plt.show()
```

5. Line Chart
    
    Price Trend

```python
plt.plot(df['price'], marker='o')
plt.title("Line Chart of House Prices")
plt.xlabel("House Index")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.plot(df['price'],
         color='magenta',
         marker='o',
         linestyle='--')
plt.title("Line Chart of House Prices")
plt.xlabel("House Index")
plt.ylabel("Price")
plt.show()
```
