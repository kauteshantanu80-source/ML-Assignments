Find mean mode and median

```python
df = pd.read_csv("Housing.csv")

def mean(data):
    return sum(data) / len(data)

def median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        mid1 = sorted_data[n // 2 - 1]
        mid2 = sorted_data[n // 2]
        return (mid1 + mid2) / 2
    else:
        return sorted_data[n // 2]

def mode(data):
    from collections import Counter
    counts = Counter(data)
    min_count=min(counts.values())
    max_count = max(counts.values())
    modes = [key for key, value in counts.items() if value == max_count]
    modes = [key for key, value in counts.items() if value == min_count]
    return modes

def variance(values):
    m = mean(values)
    total = 0
    count = 0
    for v in values:
        total += (v - m) ** 2
        count += 1
    return total / count

def data_range(values):
    return max(values)-min(values)

col = "price"

values = df[col].dropna().tolist()


# CALCULATE USING USER-DEFINED FUNCTIONS

print("USER-DEFINED RESULTS:")
print("Mean       :", mean(values))
print("Median     :", median(values))
print("Mode       :", mode(values))
print("varience:",variance(values))
print("range:",data_range(values))

# VERIFY USING INBUILT FUNCTIONS

print("\nINBUILT FUNCTION RESULTS:")
print("Mean       :", df[col].mean())
print("Median     :", df[col].median())
print("Mode       :", df[col].mode().tolist())
print("varience:",df[col].var())
print("range:",df[col].max()-df[col].min())
print("standard deviation:",df[col].std())
print("\nDESCRIBE OUTPUT:")
df.describe()
```

Vertical Data Integration (Row-wise) Combines datasets with same structure by stacking rows.

```python
import pandas as pd

jan = pd.DataFrame({
    'EmpID': [1, 2],
    'Sales': [20000, 25000]
})

feb = pd.DataFrame({
    'EmpID': [3, 4],
    'Sales': [22000, 27000]
})

# Integrate data row-wise
sales_data = pd.concat([jan, feb], ignore_index=True
                       )

print(sales_data)
```

Horizontal Data Integration (Column-wise) Combines different attributes for the same entities.
Student Academic & Attendance Data Using concat(axis=1)

```python
marks = pd.DataFrame({
    'RollNo': [101, 102, 103],
    'Marks': [85, 90, 88]
})

attendance = pd.DataFrame({
    'RollNo': [101, 102, 103],
    'Attendance': [92, 95, 90]
})

# Integrate data column-wise
student_data = pd.concat([marks, attendance['Attendance']], axis=1)

print(student_data)
```

Key-based Data Integration (Joins) Using merge() Integrates data based on a common key.


```python
students = pd.DataFrame({
    'StudentID': [1, 2, 3, 4, 5, 6],
    'Name': ['Shantanu', 'Jay', 'Aniket', 'Rushi' , 'Krushna','Apurv']
})

results = pd.DataFrame({
    'StudentID': [1,2,4,6,],
    'CGPA': [9.5, 9.1,9.7,9.1]
})

# Inner Join
data_inner = pd.merge(students, results, on='StudentID', how='inner')

print(data_inner)
```

Types of Joins in Data Integration

```python
pd.merge(students, results, on='StudentID', how='inner')   # Common records
pd.merge(students, results, on='StudentID', how='left')    # All left + matches
pd.merge(students, results, on='StudentID', how='right')   # All right + matches
pd.merge(students, results, on='StudentID', how='outer')   # All records
```

Integration Using Index Using join() Joins datasets using index instead of columns.


```python
df1 = students.set_index('StudentID')
df2 = results.set_index('StudentID')

integrated_data = df1.join(df2)

print(integrated_data)
```

Data transformation

MinMaxScaler vs StandardScaler
Both are feature scaling techniques, but they work differently.

A. MinMaxScaler

Purpose: Scales all numeric values to a fixed range, usually 0 to 1 (can be changed). Formula:

𝑋 scaled=𝑋−𝑋min/𝑋max −𝑋 min

⁡ X_min = minimum value of feature

X_max = maximum value of feature

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

df = pd.DataFrame({'Age':[15,16,17,18,19,22,25,30,35]})
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
print(df_scaled)
```

When to use:

Data is not normally distributed

You need fixed range for all features

Algorithms like Neural Networks, KNN, SVM are sensitive to scale

A. Scaling / Normalization / Standardization

Purpose: Make numerical features comparable.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

# Sample data
df = pd.DataFrame({'Age':[15,16,17,18,19,22,25,30,35], 'Salary':[2000,2500,5000,1000,20000,25000,50000,60000,80000]})

# Min-Max Normalization
scaler = MinMaxScaler()
df[['Age','Salary']] = scaler.fit_transform(df[['Age','Salary']])
print("Min-Max Scaled:\n", df)

# Standardization (Z-score)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Age','Salary']])
print("Standardized:\n", df_scaled)
```
